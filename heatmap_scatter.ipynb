```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('codesniffer.csv')
data.head()
```

```python

sns.scatterplot(data[['WCAG2A', 'WCAG2AA', 'WCAG2AAA', 'Indian Guidelines']])
plt.show()
```

```python
plt.figure(figsize=(8,6))
data['Site Link'] = data['Site Link'].astype(str)
data['WCAG2A'] = data['WCAG2A'].astype(str)
data['WCAG2AA'] = data['WCAG2AA'].astype(str)
data['WCAG2AAA'] = data['WCAG2AAA'].astype(str)
data['Indian Guidelines'] = data['Indian Guidelines'].astype(str)


# Create line chart for each score type
for score_type in ['WCAG2A', 'WCAG2AA', 'WCAG2AAA', 'Indian Guidelines']:
    plt.plot(data['Site Link'], data[score_type], label=score_type)

# Set chart title and axis labels
plt.title('Scores by Website')
plt.xlabel('Website')
plt.ylabel('Score')

# Add legend and show the chart
plt.legend()
plt.show()
```
