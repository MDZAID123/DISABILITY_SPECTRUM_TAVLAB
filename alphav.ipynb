```python
nw['WCAG2A']=nw['WCAG2A'].astype(float)
nw['WCAG2AA']=nw['WCAG2AA'].astype(float)
nw['WCAG2AAA']=nw['WCAG2AAA'].astype(float)
nw['Indian Guideline']=nw['Indian Guideline'].astype(float)
```

```python
plt.figure(figsize=(8,6))
nw['Site Link'] = nw['Site Link'].astype(str)
nw['WCAG2A'] = nw['WCAG2A'].astype(str)
nw['WCAG2AA'] =nw['WCAG2AA'].astype(str)
nw['WCAG2AAA'] = nw['WCAG2AAA'].astype(str)
nw['Indian Guideline'] =nw['Indian Guideline'].astype(str)


# Create line chart for each score type
for score_type in ['WCAG2A', 'WCAG2AA', 'WCAG2AAA', 'Indian Guideline']:
    plt.plot(nw['Site Link'], nw[score_type], label=score_type)

# Set chart title and axis labels
plt.title('Scores by Website')
plt.xlabel('Website')
plt.ylabel('Score')

# Add legend and show the chart
plt.legend()
plt.show()
```
